```python
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('fivethirtyeight')
```

```python
df1 = pd.read_csv('/home/yohan/Github/pulsar-project/tabelas/dados1.csv', delimiter=',')
df1.columns=['Tempo', 'Luminosidade']
df1[:200].plot.line(x='Tempo', y='Luminosidade', figsize=(20,10))
plt.title('Curva de Luz 1')
plt.xlabel('Tempo (milisegundos)')
plt.ylabel('Luminosidade (lux)')
plt.show()
```
