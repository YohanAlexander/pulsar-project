```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('bmh')
```

```python
df1 = pd.read_csv('tabelas/dados1.1.csv', delimiter=',')
x = df1['Tempo']
y = df1['Luz']
plt.plot(x, y)
plt.title('Curva de Luz 1')
plt.xlabel('Tempo (milisegundos)')
plt.ylabel('Luminosidade (lux)')
plt.show()
```
