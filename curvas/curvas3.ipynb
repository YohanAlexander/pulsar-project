```python
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('bmh')
```

```python
dfb = pd.read_csv('/home/yohan/Github/pulsar-project/tabelas/dadosbranco.csv', delimiter=',')
dfb.columns=['Tempo','Luminosidade']
```

```python
dfb[:400].plot.line(x='Tempo', y='Luminosidade', figsize=(20,10))
plt.title('Curva com base clara')
plt.xlabel('Tempo (milisegundos)')
plt.ylabel('Luminosidade (lux)')
plt.show()
```
