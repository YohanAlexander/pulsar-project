```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('tabelas/dados1.1.csv', delimiter=',')
x = df['Tempo']
y = df['Luz']
plt.plot(x, y)
plt.title('Curva de Luz')
plt.xlabel('Tempo (milisegundos)')
plt.ylabel('Luminosidade (lux)')
plt.show()
```
